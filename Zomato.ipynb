# 📦 Online Ordering

```python
df['online_order'].count()
```

```python
counts = df['online_order'].value_counts()

# Extract labels and values
labels = counts.index
values = counts.values

# Create pie chart
plt.figure(figsize=(4,4))
plt.pie(values, labels=labels, autopct='%1.1f%%' , startangle=90, colors=['skyblue', 'lightcoral'])
plt.title("ONLINE_ORDER - COUNT (Pie Chart)", fontdict={'fontsize':15, 'font':'Bell MT'})
plt.show()
```

#### ✅ Analysis of Online_order Count
* Restaurants that offer online ordering: ~60%
* Restaurants that not offer online ordering: ~40%

```python
plt.figure(figsize=(4,3))
ax=sns.barplot(df,x='online_order',y='rate')
ax.bar_label(ax.containers[0])
plt.title("ONLINE_ORDER vs RATE", fontdict={"fontsize":15, 'font': 'Bell MT'})
plt.show()
```

#### ✅ Analysis of average rating of online ordering

1. Average rating of restaurants with online ordering: ~3.9
2. Average rating without online ordering: ~3.5

✅ Restaurants with online ordering tend to have higher rating

```python
avg_rates = df.groupby('online_order')['votes'].mean()

# Extract labels and values
labels = avg_rates.index
values = avg_rates.values

# Create pie chart
plt.figure(figsize=(5,5))
plt.pie(values, labels=labels, autopct=lambda pct: func(pct, values), startangle=90, colors=['skyblue', 'lightcoral'])
plt.title("Online-order & Votes (Pie Chart)", fontdict={'fontsize':15, 'font':'Bell MT'})
plt.show()
```

#### ✅ Analysis of online_order and votes

* Most online-order restaurants receive higher votes compared to offline ones.

```python
plt.figure(figsize=(5,5))
sns.relplot(df,y='votes',x='approx_cost(for two people)',col='online_order')
```

#### ✅ Analysis of Online-order, Approx_cost & Votes
* Low-cost restaurants (< 300) without online ordering usually have very few votes.
* Mid-range cost restaurants (400–800) with online ordering generally get moderate-to-high votes.
* High-cost restaurants (> 800) may or may not get high votes; online ordering helps, but reputation also plays a role.

#### ✅ Analysis of Rating and Approx-cost
* Low-cost restaurants (₹100–₹300) mostly have moderate ratings (around 3.0–3.8).
* Mid-range restaurants (₹400–₹800) show higher ratings (3.8–4.4), indicating a balance between quality and affordability.
* Very high-cost restaurants (₹800–₹950) do not always have the highest ratings.

💡 Summary:
* Optimal ratings are mostly in the mid-cost range. Extreme low or high costs do not strongly correlate with higher ratings.

#### ✅ Analysis of votes and approx-cost
🔹 Votes are not strongly dependent on cost, but:
* Mid-range restaurants (₹400–₹800 for two) tend to get more votes.
* Very low-cost or very high-cost restaurants often have fewer votes.
